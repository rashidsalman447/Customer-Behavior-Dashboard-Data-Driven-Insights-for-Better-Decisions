```python
df['Review Rating']= df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median())) #to remove the missing or null value while numeric and alphbetic orders have
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()      #rearrange the columns names in snake format 
df.columns = df.columns.str.replace(' ','_')
df= df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
#create a age group
labels = ['Young Adult', 'Adult', 'Middle-Age', 'Senior']
df['age_group']= pd.qcut(df['age'], q=4, labels= labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
#create a column purchase frequency days
frequency_mapping = {'Fortnightly': 14, 'Weekly': 7, 'Montly':30, 'Quarterly': 90, 'Bi-Weekly':14, 'Annually':365, 'Every 3 Months': 90}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df[['discount_applied', 'promo_code_used']].head(10)    #we found that both column have same values
```

```python
(df['discount_applied'] == df['promo_code_used']).all()    #yes, it is true we have check and we found both the column are same
```

```python
df= df.drop( 'promo_code_used',axis = 1)   # we have remove promo_code column from table
```
